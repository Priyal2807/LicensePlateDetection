```python
from skimage.io import imread
```

```python
from skimage.filters import threshold_otsu
from skimage import measure
from skimage.measure import regionprops
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import localization
```

```python
car_image = imread("image.jpg", as_grey=True)
```

```python
gray_car_image = car_image * 255                  #multiplied with 255 as skiimage has 0 to 1 range 
fig, (ax1, ax2) = plt.subplots(1, 2)              #two plots one for grayscale image and other for black and white image
ax1.imshow(gray_car_image, cmap="gray")
threshold_value = threshold_otsu(gray_car_image)  #reading the threshold value using otsu method
binary_car_image = gray_car_image > threshold_value
ax2.imshow(binary_car_image, cmap="gray")
plt.show()
```
